```python
import pandas as pd
```

```python
product_df = pd.read_json("jsons/products.json",orient="index")

def convert_price(price: str) -> float:
    return float(price.replace(",", "."))

product_df["product_price"] = product_df["product_price"].map(convert_price)

product_df.to_json(f"JSONS/products_cleaned.json", orient="index", indent=4)
```
