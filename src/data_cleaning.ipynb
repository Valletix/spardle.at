```python
import pandas as pd
```

```python
product_df = pd.read_json("../jsons/products_combined.json",orient="index")

pd.set_option('display.max_rows', 10)

def convert_price(price: str) -> float:
    if price.startswith("statt"):
        price = price.removeprefix("statt ")
    return float(price.replace(",", "."))

def convert_name(product_name):
    if product_name == "":
        product_name = None
    return product_name


product_df["product_price"] = product_df["product_price"].map(convert_price)
product_df["product_name"] = product_df["product_name"].map(convert_name)
product_df = product_df.dropna(ignore_index=True)
product_df = product_df.drop_duplicates(ignore_index=True)
product_df.to_json(f"../jsons/products_cleaned.json", orient="index", indent=4)
```
