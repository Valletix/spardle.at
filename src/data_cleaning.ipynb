```python
import pandas as pd
```

```python
product_df = pd.read_json("../jsons/products.json",orient="index")

def convert_price(price: str) -> float:
    if price.startswith("statt"):
        price = price.removeprefix("statt ")
    return float(price.replace(",", "."))

product_df["product_price"] = product_df["product_price"].map(convert_price)

product_df.to_json(f"../jsons/products_cleaned.json", orient="index", indent=4)
```

```python
product_df = pd.read_json("../jsons/products_drinks.json",orient="index")

def convert_price(price: str) -> float:
    if price.startswith("statt"):
        price = price.removeprefix("statt ")
    return float(price.replace(",", "."))

product_df["product_price"] = product_df["product_price"].map(convert_price)

product_df.to_json(f"../jsons/products_drinks_cleaned.json", orient="index", indent=4)
```

```python
products_drinks_df = pd.read_json("../jsons/products_drinks_cleaned.json",orient="index")
products_food_df = pd.read_json("../jsons/products_food_cleaned.json",orient="index")

products_mixed_df = pd.concat([products_food_df, products_drinks_df], axis=0, ignore_index=True)

products_mixed_df.to_json(f"../jsons/products_cleaned.json", orient="index", indent=4)
```
